```
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import save_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
import zipfile
import glob
import os.path
from os import path
from random import randint
```

```
classes = ["jute", "maize", "rice", "sugarcane", "wheat"]
% cd /content/drive/MyDrive/Colab Notebooks/Computer Vision/Project/new_generated_dataset

directories = ["crop_images", "test_crop_image", "validation_crop_images"]

for directory in directories:
  os.mkdir(directory)
  for sub_class in classes:
    os.mkdir(directory + "/" +sub_class)
```

```
def displayImage(image, title=None):
  pyplot.imshow(image)
  if(title != None):
    pyplot.title(title) 
  pyplot.show()
```

```
def fetchDataGenerator(label):
  if label == 0:
    datagen = ImageDataGenerator(brightness_range=[0.2,1.0])
    return datagen
  
  if label == 1:
    datagen = ImageDataGenerator(rotation_range=90)
    return datagen
  
  if label == 2:
    datagen = ImageDataGenerator(zoom_range=[0.5,1.0])
    return datagen
```

```
for directory in directories:
  for sub_class in classes:
    paths = glob.glob("/content/drive/MyDrive/Colab Notebooks/Computer Vision/Project/original_dataset/"+ directory + "/" +sub_class+"/*.*")
    for image_path in paths:
      img = load_img(image_path)
      
      # convert to numpy array
      data = img_to_array(img)
      # expand dimension to one sample
      samples = expand_dims(data, 0)
      
      for i in range(0,3):
        datagen = fetchDataGenerator(i)
        # prepare iterator
        it = datagen.flow(samples, batch_size=1)
        # generate samples and plot
        for i in range(100):
          # pyplot.figure(figsize=(14, 14))
          # define subplot
          # pyplot.subplot(330 + 1 + i)
          # generate batch of images
          batch = it.next()
          # convert to unsigned integers for viewing
          image = batch[0].astype('uint8')

          save_img("/content/drive/MyDrive/Colab Notebooks/Computer Vision/Project/new_generated_dataset/" + directory + "/" +sub_class+"/"+sub_class+"_"+str(i)+str(randint(1, 1000000))+".jpg", image)
```
